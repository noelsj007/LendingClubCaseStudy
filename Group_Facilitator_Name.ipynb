## Bivariate Analysis:
income vs. loan amount

```python
#Removing the columns with more values are not available

threshold = 0.8
na_columns = df.columns[df.isnull().mean() > threshold]
df_cleaned = df.drop(columns = na_columns)

# standardising the funded_amnt_inv column
df_cleaned['funded_amnt_inv'] = df['funded_amnt_inv'].round().astype(int)

#conversion of terms from month to number
df_cleaned['term'] = df['term'].apply(lambda x: int(x.split()[0]))

print(df_cleaned['term'].unique())
print('Month:', df_cleaned['term'])
print('Rounded Number check:', df_cleaned.at[776, 'funded_amnt_inv'])
print('dropped Columns: ', na_columns)
```

```python
#Finding Outliers in the Loan Amount column using IQR

Q1 = df['loan_amnt'].quantile(0.25)
Q3 = df['loan_amnt'].quantile(0.75)
IQR = Q3 - Q1
print('Interquartile Range: ', IQR)

threshold = 1.5
outliers = (df['loan_amnt'] < Q1 - IQR *threshold) | (df['loan_amnt'] > Q3 + IQR * threshold)
df_cleaned = df[~outliers]

plt.bar(['original' , 'cleaned'], [len(df), len(df_cleaned)], color = ['red', 'green'])
plt.title('Difference in cleaned and original data')
plt.show()
```

## UNIVARIATE Analysis
### Business Driven analysis

```python
#Plot of people who took loan depending on the home owner ship

sns.countplot(x='home_ownership', data=df_cleaned)
plt.title('Distribution of Home Ownership Among Loan Applicants')
plt.xlabel('Home Ownership')
plt.ylabel('Count')
plt.show()

#from this it is clear that people who owns home took less loans
```

```python
#Loan term distribution analysis

plt.figure(figsize=(8, 6))
sns.countplot(x='term', data=df_cleaned.head(100))
plt.title('Loan Duration Distribution')
plt.xlabel('Loan Term (in months)')
plt.ylabel('Count')
plt.show()
```

```python
#Loan amount distribution

plt.figure(figsize=(12, 6))
plt.hist(df_cleaned['loan_amnt'], bins=30, color='skyblue', edgecolor='black')
plt.title('Loan Amount Distribution')
plt.xlabel('Loan Amount')
plt.ylabel('Count')
plt.show()
```

```python
#Loan distribution with each states

plt.figure(figsize=(15,8))
sns.countplot(x='addr_state', data=df_cleaned)
plt.title('Distribution of Loan State Among Loan Applicants')
plt.xlabel('State')
plt.ylabel('Count')
plt.show()

# from this it is identified that people from the state with code CA have took more loans
```

### Categorical Analysis

```python
#Income distribution of Loan applicants

income_ranges = [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 150000]

df_cleaned['income_range'] = pd.cut(df_cleaned['annual_inc'], bins=income_ranges)

plt.figure(figsize=(18, 6))
sns.countplot(x='income_range', color='skyblue', data=df_cleaned)
plt.title('Income Distribution of Loan Applicants')
plt.xlabel('Annual Income Range')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

#it is identified that more loan applicants are people who have income of 40000 to 60000
```

## BIVARIATE Analysis

```python
#conversion of percentage to float for analysis
df_cleaned['int_rate'] = df_cleaned['int_rate'].str.rstrip('%').astype('float')
```

```python
# Calculate the ratio between interest rate and installment

df_cleaned['int_rate_to_installment_ratio'] = df_cleaned['int_rate'].astype(int) / df_cleaned['installment']

# Scatter plot for the ratio
sns.scatterplot(x='int_rate_to_installment_ratio', y='loan_amnt', data=df_cleaned.head(100))
plt.title('Scatter Plot: Interest Rate to Installment Ratio vs. Loan Amount')
plt.xlabel('Interest Rate to Installment Ratio')
plt.ylabel('Loan Amount')
plt.show()

# form this it is clear that the interest to installment ratio is less for higher loan amounts
```

```python
#Relationship between Loan amount and terms


# Scatter plot for loan terms vs. interest rates (filtered data)
sns.scatterplot(x='loan_amnt', y='int_rate', data=df_cleaned.head(100))
plt.title('Relationship between Loan amount and terms (Filtered by Interest Rate)')
plt.xlabel('Loan term')
plt.ylabel('Interest rate')
plt.show()

# form this it is identified that for higher amount of loan the interest rate is less
```

```python
#Loan distribution graph for customers dipending on the verification status

sns.scatterplot(x='verification_status', y='loan_amnt', data=df_cleaned.head(100))
plt.title('Scatter Plot: Loan amount and Verification ratio')
plt.xlabel('Varified customer')
plt.ylabel('Loan Amount')
plt.show()

# from this it is clear that people who have a verification badge got more loan amount
```

```python
ranges = [0, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 150000]

df_cleaned['income_range'] = pd.cut(df_cleaned['annual_inc'], bins=ranges)
df_cleaned['loan_range'] = pd.cut(df_cleaned['loan_amnt'], bins=ranges)

df_cleaned['income_range'] = df_cleaned['income_range'].astype(str)
df_cleaned['loan_range'] = df_cleaned['loan_range'].astype(str)

plt.figure(figsize=(18, 12))
sns.countplot(x='income_range', hue='loan_range', data=df_cleaned)
plt.title('Income vs Loan')
plt.xlabel('Income Range')
plt.ylabel('Loan Range')
plt.xticks(rotation=45)
plt.show()

#from this it is identified that people who have income between 30000 to 40000 took more loan
#and the most approved loan amount in between 0 tp 10000
```

```python
#Committed amount and Amount committed by investors

sns.scatterplot(x='funded_amnt', y='funded_amnt_inv', data=df_cleaned.head(100))
plt.title('Scatter Plot: Committed amount and Amount committed by investors')
plt.xlabel('Committed amount')
plt.ylabel('Amount committed by investors')
plt.show()

#from this it is identified that most part of the commited amount has been funded by the investors
```
